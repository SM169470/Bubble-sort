```python
# present your code in here
my_list = [4, 2, 7, 1, 9, 5]
def bubble_sort_while(my_list):
    n = len(my_list)
    swapped = True
    while swapped:
        swapped = False
        for i in range(n - 1):
            if my_list[i] > my_list[i + 1]:
                temp = my_list[i]
                my_list[i] = my_list[i + 1]
                my_list[i + 1] = temp
                swapped = True
                print(my_list)
bubble_sort_while([4, 2, 7, 1, 9, 5])
```
